# Create a simple SGD id map - October 13, 2016

## Load the SGD features and write a map
If there is no standard name, write the systematic name again

```python
import os

data_dir = os.path.join('..', '..', 'data')
in_file = os.path.join(data_dir, 'SGD_features.tab')
out_file = os.path.join(data_dir, 'SGD_id_map.txt')

id_map = dict()

with open(in_file) as in_f:
    for line in in_f:
        parts = line.split('\t')
        assert len(parts) == 16, 'Expected 16 columns:\n{}'.format(line)
        systematic = parts[3]
        
        if systematic != '':
            standard = parts[4]

            assert systematic not in id_map, 'Already mapped {}'.format(line)

            if standard != '':
                id_map[systematic] = standard
            else:
                id_map[systematic] = systematic

print 'Loaded {} ids'.format(len(id_map))

with open(out_file, 'w') as out_f:
    for sys in sorted(id_map.keys()):
        out_f.write('{}\t{}\n'.format(sys, id_map[sys]))
```
