# Obtain list of Genes and Proteins in Osmotic stress pathway

# Import files and libraries

```python
import os.path
import matplotlib.pyplot as plt
import pandas as pd

OSMOTIC_FILEPATH = os.path.join('..', '..', 'data', 'evaluation')

data = os.path.join(OSMOTIC_FILEPATH, 'Kitano2016_data.xls')
df = pd.read_excel(data, header = 1)
print df
```

# Drop unnecessary columns

```python
df = df.drop('compartment', 1)
df = df.drop('Heat Shock', 1)
df = df.drop('Ion Homeostasis', 1)
df = df.drop('Nutrient Adapatation', 1)
df = df.drop('Oxidative Stress', 1)
df = df.drop('Pheromone Response', 1)
df = df.drop('number of occurrences', 1)
print df
```

# Obtain list of only proteins and genes

```python
df = df[df['Osmotic Stress'] != '-']
df = df[((df['class'] == 'GENE') | (df['class'] == 'PROTEIN'))]

df = df.drop('class', 1)

print df
```

```python
df = df.drop('Osmotic Stress', 1)
```

# Output to file

```python
out_file = os.path.join(OSMOTIC_FILEPATH, 'Kitano2016_Genes_Proteins.txt')

df.to_csv(out_file,  index = False, header = False, sep = '\t')
```
